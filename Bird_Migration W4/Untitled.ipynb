```python
import pandas as pd

birddata = pd.read_csv("bird_tracking.csv")
print(birddata.info())
birddata.head()
```

```python
import matplotlib.pyplot as plt
import numpy as np

bird_names = pd.unique(birddata.bird_name)
plt.figure(figsize =(7,7))
for bird_name in bird_names:
    ix = birddata.bird_name == bird_name
    x, y = birddata.longitude[ix] , birddata.latitude[ix]
    plt.plot(x,y,".", label = bird_name)
    
plt.xlabel("Longitude")
plt.ylabel("Latitude")    
plt.legend(loc="lower right")


```

```python
ix = birddata.bird_name == "Eric"
speed = birddata.speed_2d[ix]
plt.hist(speed[:10])
```

```python
print(np.isnan(speed))

print(np.isnan(speed).any())

print(np.sum(np.isnan(speed)))

ind = np.isnan(speed)
print(ind)
print(~ind)

plt.hist(speed[~ind])
plt.savefig("hist_0.pdf")
```

```python
plt.figure(figsize = (8,4))
speed = birddata.speed_2d[birddata.bird_name == "Eric"]
ind = np.isnan(speed)
plt.hist(speed[~ind], bins = np.linspace(0,30,20), normed =True)
plt.xlabel("2D speed (m/s)")
plt.ylabel("Frequency")
plt.savefig("hist_1.pdf")
```

```python
date_str = birddata.date_time[0]
print(type(date_str))

datetime.datetime.strptime(date_str[:-3], "%Y-%m-%d %H:%M:%S")

timestamps = []
for k in range(len(birddata)):
    timestamps.append(datetime.datetime.strptime\
    (birddata.date_time.iloc[k][:-3], "%Y-%m-%d %H:%M:%S"))
    
birddata["timestamp"] = pd.Series(timestamps, index = birddata.index)

birddata[birddata.bird_name == "Sanne"]
```

```python
times = birddata.timestamp[birddata.bird_name == "Eric"]
elapsed_time = [ time - times[0] for time in times]

print(elapsed_time[1000])
print(elapsed_time[1000]/ datetime.timedelta(days=1))
print(elapsed_time[1000]/ datetime.timedelta(hours=1))

plt.plot(np.array(elapsed_time)/datetime.timedelta(days=1))
plt.xlabel("Observation")
plt.ylabel("Elasped time (days)")
plt.savefig("Timeplot.pdf")
```

```python
data = birddata[birddata.bird_name == "Eric"]
times = data.timestamp
elapsed_days = np.array(elapsed_time)/datetime.timedelta(days=1)

next_day = 1
inds = []
daily_mean_speed = []
for (i, t) in enumerate(elapsed_days):
    if t < next_day:
        inds.append(i)
    else:
        #compute mean speed
        daily_mean_speed.append(np.mean(data.speed_2d[inds]))
        next_day += 1
        inds = []
        
plt.figure(figsize =( 8,6))
plt.plot(daily_mean_speed)
plt.xlabel("Days")
plt.ylabel("Mean speed in (m/s)");
plt.savefig("dms.pdf")
```

```python
#cartopy - python libratory for maps and data 

import cartopy.crs as ccrs
import cartopy.feature as cfeature

proj = ccrs.Mercator()

plt.figure(figsize =( 10,10))
ax = plt.axes(projection = proj)
ax.set_extent((-25,20,52,10)) # find by trial and errors
ax.add_feature(cfeature.LAND)
ax.add_feature(cfeature.OCEAN)
ax.add_feature(cfeature.COASTLINE)
ax.add_feature(cfeature.BORDERS, linestyle = ":")
ax.add_feature(cfeature.RIVERS)
ax.add_feature(cfeature.LAKES)

for name in bird_names:
    ix = birddata["bird_name"] == name
    x, y = birddata.longitude[ix] , birddata.latitude[ix]
    ax.plot(x,y,".", transform=ccrs.Geodetic(), label = name)
    
plt.legend(loc ="upper left")
plt.savefig("map.pdf")
```
